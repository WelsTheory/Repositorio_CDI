```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
muestra = pd.read_csv('Data_Nucleo2.csv')
```

```python
SAMPLES = 1000

SAMPLE_INDEX = 0
plt.figure(figsize=(20, 5))
plt.plot(muestra['TIEMPO'][SAMPLE_INDEX:SAMPLES], muestra['ADC'][SAMPLE_INDEX:SAMPLES],'-', label='ADC')
plt.plot(muestra['TIEMPO'][SAMPLE_INDEX:SAMPLES], muestra['DAC'][SAMPLE_INDEX:SAMPLES], '-',label='DAC')
plt.title('System Pulse Response')
plt.ylabel('Magnitude')
plt.xlabel('Time [Sec]')
plt.legend()
plt.grid()
```

```python
# Encontrar el ndice donde la ADC comienza a subir
inicio_subida = muestra['ADC'].gt(0.1).idxmax()

# Encontrar el ndice donde la ADC alcanza el valor mximo (1)
fin_subida = muestra['ADC'].gt(0.9).idxmax()

# Calcular el tiempo de subida en segundos
tiempo_subida = muestra.loc[fin_subida, 'TIEMPO'] - muestra.loc[inicio_subida, 'TIEMPO']

print("El tiempo de subida es:", tiempo_subida, "segundos")
```
